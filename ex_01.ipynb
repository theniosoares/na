R: Um grafo é chamado hamiltoniano se existe nele um ciclo hamiltoniano , ou seja, existe um ciclo que contenha todos os vértices, sendo que cada vértice só aparece uma vez no ciclo.

R: Um grafo é chamado euleriano se existe nele um ciclo euleriano, ou seja, existe um ciclo que contenha todos as arestas.

```python
### Questão 6
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([('M', 'S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'), ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')])
nx.draw(G, with_labels=True)
plt.show()
```

```python
#### Questão 7

G = nx.Graph()
G.add_edges_from([(2, 5), (6, 1), (5, 3), (2, 3)])
nx.draw(G, with_labels=True)
plt.show()

G = nx.DiGraph()
G.add_edges_from([(2, 5), (6, 1), (5, 3), (2, 3)])
nx.draw(G, with_labels=True)
plt.show()
```

```python
### Questão 9

# Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship
# with both C and D, but only C considers D a friend. D has also sent friend requests 
# to E, F, G, and H but, so far, only G replied. G also has a reciprocal relationship with A. 
# Draw the corresponding directed graph.

G = nx.DiGraph()
G.add_edges_from([('A', 'B'), ('A', 'G'), ('B', 'C'), ('B', 'D'), ('C', 'B'), ('C', 'D'), ('D', 'B'), ('D', 'E'), ('D', 'F'), ('D', 'G'), ('D', 'H'), ('G', 'D'), ('G', 'A')])
nx.draw(G, with_labels=True)
plt.show()
```
