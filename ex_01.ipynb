R: Não existe nenhum ciclo simples envolvendo ao menos 4 vértices.

(c) Is the digraph connected or not connected?
R: O grafo é conectado.

(e) Represent the digraph using adjacency list representation.
R: M[S], N[R,Q,O], O[M,N], P[P,R,M], Q[], R[], S[M]

(f) Represent the digraph using adjacency matrix representation. R:\
M N O P Q R S \

N [0, 0, 1, 0, 1, 1, 0]\
O [1, 1, 0, 0, 0, 0, 0]\

```python
### Questão 6

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([('M', 'S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'), ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')])
nx.draw(G, with_labels=True)
plt.show()
```
